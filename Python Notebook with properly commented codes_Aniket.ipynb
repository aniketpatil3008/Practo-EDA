### Data Cleaning and Validation

```python
# Converting categorical/indicator variables to 'category' datatype
categorical_columns = ['HeartDiseaseorAttack', 'HighBP', 'HighChol', 'CholCheck', 'Smoker', 'Stroke',
                       'Diabetes', 'PhysActivity', 'Fruits', 'Veggies', 'HvyAlcoholConsump',
                       'AnyHealthcare', 'NoDocbcCost', 'DiffWalk', 'Sex', 'Age', 'Education', 'Income']

data[categorical_columns] = data[categorical_columns].astype('category')

# Validate the data types after conversion
data.info()
```

### Univariate Analysis

```python
# Set plot style for better readability
sns.set(style="whitegrid")
```

```python
# Plot distribution of BMI
plt.figure(figsize=(10, 6))
sns.histplot(data['BMI'], bins=30, kde=True, color='blue')
plt.title('Distribution of BMI')
plt.xlabel('BMI')
plt.ylabel('Frequency')
plt.show()
```

```python
# Define mappings for Age and Education
age_mapping = {
    1: '18-24', 2: '25-29', 3: '30-34', 4: '35-39', 5: '40-44', 6: '45-49',
    7: '50-54', 8: '55-59', 9: '60-64', 10: '65-69', 11: '70-74', 12: '75-79', 13: '80 or older'
}

education_mapping = {
    1: 'Never attended school', 2: 'Elementary school', 3: 'Some high school',
    4: 'High school graduate', 5: 'Some college', 6: 'College graduate'
}

# Apply mappings to create new columns with readable labels
data['Age Group'] = data['Age'].map(age_mapping)
data['Education Level'] = data['Education'].map(education_mapping)
```

```python
# Plot distribution of Age categories
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='Age', palette='Set2')
plt.title('Distribution of Age Groups')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.show()
```

```python
# Plot distribution of Mental Health Days (MentHlth)
plt.figure(figsize=(10, 6))
sns.histplot(data['MentHlth'], bins=30, kde=True, color='green')
plt.title('Distribution of Mental Health Days')
plt.xlabel('Number of MentHlth Days')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plot distribution of Physical Health Days (PhysHlth)
plt.figure(figsize=(10, 6))
sns.histplot(data['PhysHlth'], bins=30, kde=True, color='orange')
plt.title('Distribution of Physical Health Days')
plt.xlabel('Number of PhysHlth Days')
plt.ylabel('Frequency')
plt.show()
```

```python
# Bar chart for prevalence of High Blood Pressure (HighBP)
plt.figure(figsize=(8, 5))
sns.countplot(data=data, x='HighBP', palette='Set1')
plt.title('Prevalence of High Blood Pressure')
plt.xlabel('HighBP (0 = No, 1 = Yes)')
plt.ylabel('Count')
plt.show()
```

```python
# Bar chart for prevalence of High Cholesterol (HighChol)
plt.figure(figsize=(8, 5))
sns.countplot(data=data, x='HighChol', palette='Set2')
plt.title('Prevalence of High Cholesterol')
plt.xlabel('HighChol (0 = No, 1 = Yes)')
plt.ylabel('Count')
plt.show()
```

```python
# Bar chart for prevalence of Stroke
plt.figure(figsize=(8, 5))
sns.countplot(data=data, x='Stroke', palette='Set3')
plt.title('Prevalence of Stroke')
plt.xlabel('Stroke (0 = No, 1 = Yes)')
plt.ylabel('Count')
plt.show()
```

```python
# Bar chart for prevalence of Diabetes
plt.figure(figsize=(8, 5))
sns.countplot(data=data, x='Diabetes', palette='Set3')
plt.title('Prevalence of Diabetes')
plt.xlabel('Diabetes (0 = No, 1 = Yes)')
plt.ylabel('Count')
plt.show()
```

```python
# Bar chart for distribution of Heart Disease
plt.figure(figsize=(8, 5))
sns.countplot(data=data, x='HeartDiseaseorAttack', palette='Set1')
plt.title('Distribution of Heart Disease Cases')
plt.xlabel('Heart Disease (0 = No, 1 = Yes)')
plt.ylabel('Count')
plt.show()
```

```python
# Pie chart for Heart Disease distribution
data['HeartDiseaseorAttack'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90, colors=['lightblue', 'orange'], figsize=(6, 6))
plt.title('Heart Disease Distribution')
plt.ylabel('')
plt.show()
```

### Bivariate Analysis

```python
# Compare HighBP distribution with Heart Disease
plt.figure(figsize=(8, 5))
sns.countplot(data=data, x='HighBP', hue='HeartDiseaseorAttack', palette='Set2')
plt.title('HighBP vs Heart Disease')
plt.xlabel('HighBP (0 = No, 1 = Yes)')
plt.ylabel('Count')
plt.show()
```

```python
# Compare HighChol distribution with Heart Disease
plt.figure(figsize=(8, 5))
sns.countplot(data=data, x='HighChol', hue='HeartDiseaseorAttack', palette='Set3')
plt.title('HighChol vs Heart Disease')
plt.xlabel('HighChol (0 = No, 1 = Yes)')
plt.ylabel('Count')
plt.show()
```

```python
# BMI distribution for individuals with and without heart disease
plt.figure(figsize=(10, 6))
sns.boxplot(data=data, x='HeartDiseaseorAttack', y='BMI', palette='Set1')
plt.title('BMI vs Heart Disease')
plt.xlabel('Heart Disease (0 = No, 1 = Yes)')
plt.ylabel('BMI')
plt.show()
```

### Correlation Matrix

```python
# Compute correlation matrix for continuous variables
corr_matrix = data[['BMI', 'MentHlth', 'PhysHlth']].corr()

# Plot the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix of Continuous Variables')
plt.show()
```
